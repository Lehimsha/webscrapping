```python
import requests
```

```python
import bs4
```

```python
url="https://www.bikewale.com/dealer-showrooms/royalenfield/mumbai/"
```

```python
data=requests.get(url)
data
```

```python
data.text
type(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=bs4.BeautifulSoup(data.text,'html.parser')
```

```python
soup.find_all()
```

```python
soup.find('h3',attrs={'class':"list-item__showroom-name"}).text
```

```python
soup.find('p',attrs={'class':"list-item__showroom-location"}).text
```

```python
soup.find('p',attrs={'class':"list-item__showroom-contact"}).text
```

```python
l=soup.find_all('h3',attrs={'class':"list-item__showroom-name"})
```

```python
name=[]
for i in l:
    name.append(i.text)
```

```python
print(len(l))    
```

```python
name=[i.text for i in l]    
print(name)
```

```python
x=soup.find_all('p',attrs={'class':"list-item__showroom-location"})
locations=[i.text.replace("\n","") for i in x]
print(locations)
print(len(locations))
```

```python
y=soup.find_all('p',attrs={'class':"list-item__showroom-contact"})
z=[i.text for i in y]
print(z)
z=[i.text for i in y]
print("9am to 7pm")
```

```python
import pandas as pd
```

```python
data=pd.DataFrame({"branch_name":name,"location_coordinates":locations,"contact":z,"timings":"9am to 7pm"})
data
```

```python
data.to_csv("RE1.csv")
```
